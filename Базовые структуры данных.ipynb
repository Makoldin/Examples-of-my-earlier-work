Программа должна запрашивать количество секунд S, которое прошло с начала суток, и выводить в формате h:mm:ss, какое время будут показывать часы. <br><br>

*Формат ввода:* <br>
3600<br><br>

*Формат вывода:* <br>
1:00:00

```python
n = int(input())
sec = (n // 1) % 60
minutes = (n // 60) % 60
hour = int((n // 3600) % 24)
t = str(hour)+ ':' + str(minutes)+':' + str(sec)
t
```
